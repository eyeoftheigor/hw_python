```python
import pandas as pd
import re

# 1. Считываем датасет
df = pd.read_csv("train.csv")

# 2. Основная информация о датасете
print("=== Первые строки датасета ===")
print(df.head(), "\n")

print("=== Информация о типах данных ===")
print(df.info(), "\n")

print("=== Количество пропусков ===")
print(df.isnull().sum(), "\n")

print("=== Описательная статистика ===")
print(df.describe(include="all"), "\n")

# 3. Процент выживаемости у каждого класса пассажиров
survival_rate = df.groupby("Pclass")["Survived"].mean() * 100
print("=== Процент выживаемости по классам ===")
print(survival_rate, "\n")


# --- Функция извлечения имени ---
def extract_first_name(name):
    # Формат: LastName, Title. FirstName ...
    match = re.search(r",\s*[^.]+\.\s*([A-Za-z]+)", name)
    if match:
        return match.group(1)
    return None


df["FirstName"] = df["Name"].apply(extract_first_name)

# 4. Самое популярное мужское и женское имя
male_name = df[df["Sex"] == "male"]["FirstName"].value_counts().idxmax()
female_name = df[df["Sex"] == "female"]["FirstName"].value_counts().idxmax()

print("=== Самые популярные имена ===")
print("Мужское:", male_name)
print("Женское:", female_name, "\n")


# 5. Самые популярные имена по полу в каждом классе
print("=== Самые популярные имена по полу в каждом классе ===")

for pclass in sorted(df["Pclass"].unique()):
    class_df = df[df["Pclass"] == pclass]

    male_top = class_df[class_df["Sex"] == "male"]["FirstName"].value_counts()
    female_top = class_df[class_df["Sex"] == "female"]["FirstName"].value_counts()

    male_top = male_top.idxmax() if not male_top.empty else None
    female_top = female_top.idxmax() if not female_top.empty else None

    print(f"Pclass {pclass}:")
    print("  Мужское имя:", male_top)
    print("  Женское имя:", female_top)

print()


# 6. Пассажиры старше 44 лет
print("=== Пассажиры старше 44 лет ===")
print(df[df["Age"] > 44].head(), "\n")


# 7. Пассажиры младше 44 лет и мужского пола
print("=== Мужчины младше 44 лет ===")
print(df[(df["Age"] < 44) & (df["Sex"] == "male")].head(), "\n")


# 8. Количество n-местных кабин
# В поле Cabin может быть несколько кабин через пробел
cabins = df["Cabin"].dropna().str.split()
cabins = cabins.explode()

# Сколько пассажиров в каждой конкретной кабине
cabin_sizes = cabins.value_counts()

# Сколько кабин на 1,2,3,... человек
n_person_cabins = cabin_sizes.value_counts().sort_index()

print("=== Количество n-местных кабин ===")
print(n_person_cabins, "\n")


# 9. Количество пассажиров без родственников на борту
# SibSp — супруги/братья/сестры, Parch — родители/дети
no_relatives = df[(df["SibSp"] + df["Parch"]) == 0].shape[0]

print("=== Пассажиры без родственников на борту ===")
print(no_relatives)
```
